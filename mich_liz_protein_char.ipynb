#### Analyzing the AA Sequences & Adding additional Characteristics

```python
# applying the Protein Analysis function of BioPython to convert
# amino acids to be evaluated though built in functions (molecular weight)
analyze_prot = df_clean_fin["sequence"].apply(ProteinAnalysis)
```

```python
# GOOD LUCK :D
list(analyze_prot)[0].count_amino_acids()['A']
```

```python
# testing to see if molecular weight is working on a smaller dataset
list_prot = list(analyze_prot[:10])
molecular_w = [x.molecular_weight() for x in list_prot]
merge_list = tuple(zip(list(prot_seq["sequence"][:10]), molecular_w))
merge_list
```

```python
list_prot = list(analyze_prot)

df_prot_char = df_clean_fin.copy()
df_prot_char["seq_len"] = df_clean_fin['sequence'].apply(len) 
df_prot_char["molecular_weight"] = [x.molecular_weight() for x in list_prot]
df_prot_char
```

```python
df_prot_char["hydrophobicity" ] = [x.gravy() for x in list_prot]
df_prot_char["instability_index"] = [x.instability_index() for x in list_prot]
df_prot_char["aromaticity"] = [x.aromaticity() for x in list_prot]
```
