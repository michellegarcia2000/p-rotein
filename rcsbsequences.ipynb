```python
# create list of json sequences
from os import listdir
#strpath = "C:\Users\kathe\Downloads\p-ai\rcsb_sequences" # this is wrong
strpath = "/Users/kathe/Downloads/p-ai/rcsb_sequences"

lstseq = listdir(strpath)
lstpath = ["/Users/kathe/Downloads/p-ai/rcsb_sequences/" + x for x in lstseq]
lstpath[1]
```

```python
# https://stackoverflow.com/questions/57067551/how-to-read-multiple-json-files-into-pandas-dataframe
# This worked! using the list of files above, make a bunch of dfs and then append
# actually someone in stackoverflow says "DataFrame.append inside a loop because it leads to quadratic copying" so we might have to fix that but it works for now
dfs = []
file_list = lstpath
for file in file_list:
    with open(file) as f:
        #json_data = pd.json_normalize(json.loads(f.read())) #this doesn't normalize correctly
        #json_data = pd.json_normalize(json_data, record_path=['data', 'polymer_entities'], meta=[['data', 'rcsb_id'], 'identifier']) # this uses wrong first argument
        json_data = pd.json_normalize(json.loads(f.read()), record_path=['data', 'polymer_entities'], meta=[['data', 'rcsb_id'], 'identifier'])
    dfs.append(json_data)
df = pd.concat(dfs, sort=False) # or sort=True depending on your needs
df
```
