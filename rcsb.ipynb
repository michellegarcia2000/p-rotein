```python
import json
import requests
import pandas as pd
import os
```

```python
strpathlocal = "./rcsb_sequences/"

lstseqlocal = os.listdir(strpathlocal)
lstpathlocal = ["./rcsb_sequences/" + x for x in lstseqlocal]
lstpathlocal[1]
```

```python
dfs = []
file_list = lstpathlocal
for file in file_list:

    with open(file) as f:
        json_data = pd.json_normalize(json.loads(f.read()), record_path=['data', 'polymer_entities'], meta=[['data', 'rcsb_id'], 'identifier'])
    dfs.append(json_data)
df = pd.concat(dfs, sort=False)
```
